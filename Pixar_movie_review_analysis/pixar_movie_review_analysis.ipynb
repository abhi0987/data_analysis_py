```python
#import the library

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
%matplotlib inline
import pandas as pd
from pandas import DataFrame,Series
```

```python
pixar_df = pd.read_csv('E:/Github_dir/data_analysis_py/Pixar_movie_review_analysis/PixarMovies.csv')
pixar_df
```

```python
# get data types of each column
pixar_df.dtypes
```

```python
pixar_df=pixar_df.dropna()
pixar_df
```

```python
pixar_df['Domestic %'] = pixar_df['Domestic %'].str.rstrip('%').astype(float)
pixar_df['International %'] = pixar_df['International %'].str.rstrip('%').astype(float)

pixar_df.loc[:,['IMDB Score']] = pixar_df['IMDB Score'] * 10

pixar_df
```

```python
# visualization

pixar_df.plot(x = 'Movie', y = ['RT Score','IMDB Score','Metacritic Score'])
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sn.barplot(x='Movie',y='IMDB Score',data=pixar_df)
```

```python
#Check range of reviews using box plot

revZ_mov = pixar_df[['RT Score','IMDB Score','Metacritic Score']]
revZ_mov.plot(kind='box',figsize=(10,6))
```

```python
plt.figure(figsize=(20, 6))
revnue_properties = pixar_df[['Movie','Domestic %','International %']]
revnue_properties.plot(x='Movie', kind='bar',stacked=True)
```
