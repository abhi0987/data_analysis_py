```python
test_df['Age'] = test_df['Age'].fillna(df['Age'].median())
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())
embarked_mapping_ = {"S": 0, "C": 1,"Q":2}
test_df['Embarked'] = test_df['Embarked'].map(embarked_mapping_).astype(int)
test_df['Embarked'].fillna(0,inplace=True)



#Creating person column in  test data
test_df['Person'] = test_df[['Age','Sex']].apply(gender_checker,axis=1) 

gender_mapping = {"female": 1, "male": 0,"Child": 2}
test_df['Person'] = test_df['Person'].map(gender_mapping).astype(int)




test_df['Alone'] = test_df.SibSp + test_df.Parch
test_df['Alone'].loc[test_df['Alone']>0]=1
test_df['Alone'].loc[test_df['Alone']==0]=0

for data in test_df:
    test_df['Title'] = test_df.Name.str.extract(' ([A-Za-z]+)\.',expand=False)

pd.crosstab(test_df['Title'],test_df['Sex'])

rare = ['Capt','Col','Countess','Don','Dr','Jonkheer','Lady','Major','Rev', 'Sir']   
test_df['Title'] = test_df['Title'].replace(rare,'Rare')

# Some other conversion
test_df['Title'] = test_df['Title'].replace('Mlle','Miss')
test_df['Title'] = test_df['Title'].replace('Ms','Miss')
test_df['Title'] = test_df['Title'].replace('Mme','Mrs')

title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
test_df['Title'] = test_df['Title'].map(title_mapping)
test_df['Title'] = test_df['Title'].fillna(0)




test_df['Title'] = test_df['Title'].astype(int)





gender_mapping = {"female": 1, "male": 0}
test_df['Sex'] = test_df['Sex'].map(gender_mapping).astype(int)


df = df.drop(['Parch','SibSp'],axis=1)
test_df = test_df.drop(['Parch','SibSp'],axis=1)

```

```python
df['Age'] = df['Age'].astype(int)
test_df['Age'] = test_df['Age'].astype(int)

#Create age bands

df['AgeBand'] = pd.cut(df['Age'],5)
df[["AgeBand","Survived"]].groupby("AgeBand",as_index=False).mean().sort_values(by="AgeBand",ascending=True)
    
df.loc[ df['Age'] <= 16, 'Age'] = 0
df.loc[(df['Age'] > 16) & (df['Age'] <= 32), 'Age'] = 1
df.loc[(df['Age'] > 32) & (df['Age'] <= 48), 'Age'] = 2
df.loc[(df['Age'] > 48) & (df['Age'] <= 64), 'Age'] = 3
df.loc[ df['Age'] > 64, 'Age'] = 4


test_df.loc[ test_df['Age'] <= 16, 'Age'] = 0
test_df.loc[(test_df['Age'] > 16) & (test_df['Age'] <= 32), 'Age'] = 1
test_df.loc[(test_df['Age'] > 32) & (test_df['Age'] <= 48), 'Age'] = 2
test_df.loc[(test_df['Age'] > 48) & (test_df['Age'] <= 64), 'Age'] = 3
test_df.loc[ test_df['Age'] > 64, 'Age'] = 4

df = df.drop('AgeBand',axis=1)
```

```python
combine = [df,test_df]
```

```python
dummy0 = pd.get_dummies(combine[0].Pclass,prefix='Pclass')
dummy1 = pd.get_dummies(combine[1].Pclass,prefix='Pclass')
    
combine[0] = pd.concat([combine[0],dummy0],axis=1)
combine[1] = pd.concat([combine[1],dummy1],axis=1)
    

combine[1].head()    
```

```python
dummy0 = pd.get_dummies(combine[0].Embarked,prefix='Embarked')
dummy1 = pd.get_dummies(combine[1].Embarked,prefix='Embarked')
    
combine[0] = pd.concat([combine[0],dummy0],axis=1)
combine[1] = pd.concat([combine[1],dummy1],axis=1)
```

```python
#These are the predictive columns 
predictors_cols = ["Pclass_1","Pclass_2","Pclass_3", "Age", "Fare", "Embarked_0","Embarked_1","Embarked_2" , "Sex" , "Alone", "Person" , "Title"]
```

```python
# Let's Go for predictive model
# We will use Logistic regerssion for prediction
#Random Forest Classifier
#KNeighbourClassifier
from sklearn import cross_validation as cv
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier

# initialize algorithim
algo_ = LogisticRegression(random_state=1)
```

```python
df = combine[0]
test_df = combine[1]
```

```python
algo = LogisticRegression(random_state=1)
X_train = df[predictors_cols]
Y_train = df['Survived']
X_test = test_df[predictors_cols]
#train the algorithim using all the train data
algo.fit(X_train,Y_train)
predictions = algo.predict(X_test)


acc_log = round(algo.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
# This is  Support Vector Machine Algorithim
svc = SVC()
svc.fit(X_train,Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train,Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

```python

# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred_rand = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```
