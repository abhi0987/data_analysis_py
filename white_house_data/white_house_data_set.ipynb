# Calculating length of each position title

```python
dc_df['Length of job title'] = dc_df['Position Title'].apply(lambda x : len(x))

dc_df.head(20)
```

```python
dc_df['Length of job title'].value_counts()
```

# Create a column title - lengh - range to put the values in a range

```python
def title_length_category(data):
    length = data['Length of job title']
    if 0 < length <=20 :
        return '0-20'
    if 20 < length <=40 :
        return '20-40'
    if 40 < length <=60 :
        return '40-60'
    if 60 < length <=80 :
        return '60-80'
    if 80 < length <=100 :
        return '80-100'
    if 100 < length <=120 :
        return '100-120'
    if 120 < length <=140 :
        return '120-140'
    if 140 < length <=160 :
        return '140-160'

dc_df['Title-length-range'] =  dc_df.apply(title_length_category,axis=1) 
```

```python
dc_df.head(10)
```

# Plot title-length-range vs Salary using seaborn

```python
plot = sns.factorplot(x='Title-length-range',y='Salary',data = dc_df,kind='bar',palette='cool',size =10.0,aspect=1.6)
plot.set_xticklabels(rotation=30)
```
